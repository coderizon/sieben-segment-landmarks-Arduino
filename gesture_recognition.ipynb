```python
import os
import tensorflow as tf
assert tf.__version__.startswith('2')

from mediapipe_model_maker import gesture_recognizer
import matplotlib.pyplot as plt

print("Imports erfolgreich geladen.")
```

```python
dataset_path = "rps_data_sample"
print(f"Daten liegen im Ordner: {dataset_path}")
```

```python
print(f"Datensatz-Pfad: {dataset_path}")

labels = []
for i in os.listdir(dataset_path):
  if os.path.isdir(os.path.join(dataset_path, i)):
    labels.append(i)

print(f"Gefundene Labels: {labels}")
```

```python
# Daten laden und vorverarbeiten
# Hinweis: MediaPipe luft jetzt ber jedes Bild und extrahiert Hand-Landmarks.
data = gesture_recognizer.Dataset.from_folder(
    dirname=dataset_path,
    hparams=gesture_recognizer.HandDataPreprocessingParams()
)

# Aufteilen in Training (80%), Validierung (10%) und Test (10%)
train_data, rest_data = data.split(0.8)
validation_data, test_data = rest_data.split(0.5)

print("Daten erfolgreich geladen und aufgeteilt.")
```

```python
# Hyperparameter definieren (Wo soll das Modell gespeichert werden?)
hparams = gesture_recognizer.HParams(export_dir="exported_model")

# Optionen zusammenstellen
options = gesture_recognizer.GestureRecognizerOptions(hparams=hparams)

print("Starte Training...")

# Das eigentliche Training starten
model = gesture_recognizer.GestureRecognizer.create(
    train_data=train_data,
    validation_data=validation_data,
    options=options
)

print("Training abgeschlossen.")
```

```python
# Testen mit den beiseite gelegten Daten
loss, acc = model.evaluate(test_data, batch_size=1)
print(f"Test loss: {loss}, Test accuracy: {acc}")
```

```python
# Modell in den Ordner 'exported_model' schreiben
model.export_model()

# Inhalt des Ordners anzeigen, um zu prfen, ob die Datei da ist
print("Exportierte Dateien:")
!ls -lh exported_model
```
